```python
def flat_gen(list1):
    if type(list1) is list:
        for i in list1:
            yield from flat_gen(i)
    else:
        yield list1

def flatten(list1):
    return list(flat_gen(list1))

list1 = [[1,'a',['cat'],2],[[[3]],'dog'],4,5]
flat_list = flatten(list1)
print(flat_list)
```
